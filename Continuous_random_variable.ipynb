```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sp
```

```python
x = np.linspace(0, 10, 100)
a, b= 3, 4
y = sp.uniform.pdf(x, a, b)
plt.plot(x, y, "b")
```

```python
x = sp.norm.rvs(0, 1, 1000)
pdf = sp.norm.pdf(np.sort(x), 0, 1)
plt.plot(np.sort(x), pdf)
plt.hist(x, density=True)
plt.show()
```

```python
import seaborn as scipy
from scipy.stats import expon
```

```python
x = np.linspace(0.001, 10, 1000)
plt.plot(x, (expon.pdf(x, 0, 2)))
plt.plot(x, (expon.pdf(x, 0, 1)))
plt.plot(x, (expon.pdf(x, 0, 0.5)))
```

```python
x = np.linspace(0, 10, 100)
g1 = sp.gamma.pdf(x, a=2, scale=1/3)
g2 = sp.gamma.pdf(x, a=1, scale=1)
g3 = sp.gamma.pdf(x, a=2, scale=2)
g4 = sp.gamma.pdf(x, a=2, scale=1)

plt.plot(x, g1, label=("alpha=2, beta = 1/3"))
plt.plot(x, g2, label=("alpha=1, beta = 1"))
plt.plot(x, g3, label=("alpha=2, beta = 2"))
plt.plot(x, g4, label=("alpha=2, beta = 1"))

plt.legend()
plt.show()
```
